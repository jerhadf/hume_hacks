```python
import requests
import io
from IPython.display import Audio

# We can some endpoints up in test cluster (https://test-api.hume.ai), this might be helpful for frontend / design (
# @Mary Jiang
 
# @Zach Krall
# )
# GET /v0/speakers -> List hosted speakers
# GET /v0/speakers/sam -> Get a speaker (by name)
# GET /v0/speakers/sam/samples -> List samples for a speaker
# GET /v0/speakers/sam/samples/1 -> Get a sample by ID
# GET /v0/speakers/sam/samples/1/url -> Get a signed URL for a sample
```

```python
HUME_API_KEY = input("Please enter your Hume API key: ")
```

```python
def list_hosted_speakers(api_key: str) -> None:
    """
    Sends a GET request to the Hume API to list the currently hosted speakers.
    """
    url = "https://test-api.hume.ai/v0/speakers"
    headers = {
        "X-Hume-Api-Key": api_key,
    }

    # Send the GET request
    response = requests.get(url, headers=headers)

    # Print the response content
    print(response.json())
    
current_speakers = list_hosted_speakers(HUME_API_KEY)
```

```python

def get_speaker_by_name(api_key: str, speaker_name: str) -> None:
    """
    Sends a GET request to the Hume API to get a speaker by name.
    """
    url = f"https://test-api.hume.ai/v0/speakers/{speaker_name}"
    headers = {
        "X-Hume-Api-Key": api_key,
    }

    # Send the GET request
    response = requests.get(url, headers=headers)

    # Print the response content
    print(response.json())

def list_samples_for_speaker(api_key: str, speaker_name: str) -> None:
    """
    Sends a GET request to the Hume API to list samples for a speaker.
    """
    url = f"https://test-api.hume.ai/v0/speakers/{speaker_name}/samples"
    headers = {
        "X-Hume-Api-Key": api_key,
    }

    # Send the GET request
    response = requests.get(url, headers=headers)

    # Print the response content
    print(response.json())

def get_sample_by_id(api_key: str, speaker_name: str, sample_id: int) -> None:
    """
    Sends a GET request to the Hume API to get a sample by ID.
    """
    url = f"https://test-api.hume.ai/v0/speakers/{speaker_name}/samples/{sample_id}"
    headers = {
        "X-Hume-Api-Key": api_key,
    }

    # Send the GET request
    response = requests.get(url, headers=headers)

    # Print the response content
    print(response.json())

def get_signed_url_for_sample(api_key: str, speaker_name: str, sample_id: int) -> None:
    """
    Sends a GET request to the Hume API to get a signed URL for a sample.
    """
    url = f"https://test-api.hume.ai/v0/speakers/{speaker_name}/samples/{sample_id}/url"
    headers = {
        "X-Hume-Api-Key": api_key,
    }

    # Send the GET request
    response = requests.get(url, headers=headers)

    # Print the response content
    print(response.json())


```

```python
def send_TTS_request(text: str, reference_speaker: str, reference_sample: int, api_key: str) -> None:
    """
    Sends a POST request to the Hume API to generate audio and plays the audio.

    Args:
        text (str): The text to be converted to speech.
        reference_speaker (str): The reference speaker for the speech.
        reference_sample (int): The reference sample for the speech.
        api_key (str): The API key for the Hume API.

    Returns:
        None
    """
    url = "https://test-api.hume.ai/v0/tts"
    data = {
        "text": text,
        "reference_speaker": reference_speaker,
        "reference_sample": reference_sample
    }
    headers = {
        "X-Hume-Api-Key": api_key,
        "Content-Type": "application/json"
    }

    # Send the POST request
    response = requests.post(url, json=data, headers=headers)

    # Play the response content as audio
    audio = Audio(response.content, autoplay=True)
    return audio

text = "Which way to the train station?"
reference_speaker = "sam"
reference_sample = 3

send_TTS_request(text, reference_speaker, reference_sample, HUME_API_KEY)
```
