```python
# Get the latest created prompts
response = requests.get(
    f"{API_BASE_URL}/prompts",
    headers=headers
)

prompts = response.json()

print(prompts)

if "prompts_page" in prompts:
    latest_prompt = prompts["prompts_page"][0]
    prompt_id = latest_prompt["id"]
    print(f"Latest prompt ID: {prompt_id}")
else:
    print("No prompts found or unexpected response format.")
    prompt_id = None
```

```python
# Step 2: Create the config
config_data = {
    "name": "Hume Documentation Assistant Configuration",
    "prompt": {
        "id": prompt_id,
    },
    "voice": {
        "provider": "HUME_AI",
        "name": "ITO"
    },
    "language_model": {
        "model_provider": "ANTHROPIC",
        "model_resource": "claude-3-sonnet-20240229",
        "temperature": 0.7
    },
    "builtin_tools": [
        {
            "name": "web_search",
            "fallback_content": "Sorry, but I couldn't find any relevant information in the Hume AI developer documentation at dev.hume.ai. Could you please rephrase your question or ask about a different topic related to Hume AI's APIs or services?"
        }
    ],
    "event_messages": {
        "on_new_chat": {
            "enabled": True,
            "text": "Hello! I'm an AI assistant specialized in helping with Hume AI developer documentation on dev.hume.ai. What would you like to know about Hume's APIs?"
        }
    },
    "timeouts": {
        "inactivity": {
            "enabled": True,
            "duration_secs": 300
        },
        "max_duration": {
            "enabled": True,
            "duration_secs": 1800
        }
    }
}

config_response = requests.post(f"{API_BASE_URL}/configs", json=config_data, headers=headers)
print(f"Config response: {config_response.json()}")
config_id = config_response.json()["id"]

print(f"Configuration created successfully. Config ID: {config_id}")
```
